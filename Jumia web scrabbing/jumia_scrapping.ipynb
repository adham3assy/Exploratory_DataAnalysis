```python
from selenium import webdriver
import pandas as pd
import os
%pip install requests
import requests
```

```python
adham = webdriver.Chrome()
adham.get("https://www.jumia.com.eg/playstation-5/#catalog-listing")
```

```python
products= adham.find_elements("xpath", '//div[@class="info"]')

```

```python
[p.text.split("\n") for p in products]
```

```python
names = []
rates = []
prices = []
discounts = []

for p in products:
    info = p.text.split("\n")
    names.append(info[0])
    prices.append(info[1])
    
    discount_found = False
    rate_found = False
    for i in info:
        if "%" in i:
            discounts.append(i)
            discount_found = True

        
        if "out of 5" in i:
            rates.append(i)
            rate_found = True
    if not rate_found:
        rates.append("None")

    if not discount_found:
        discounts.append("None")

        
        
```

```python
images = adham.find_elements("xpath", '//div[@class="img-c"]/img[@class="img"]')
images = [i.get_attribute("data-src") for i in images]
images
```

```python
data = {"name":names, "price":prices, "discount":discounts, "rate":rates , "image":images}
df = pd.DataFrame(data)
df
```

```python
df.to_csv("jumia.csv", index=False)
```

```python
for index, url in enumerate(images, start=1):
    try:
        response = requests.get(url, timeout=10)  
        response.raise_for_status()

        ext = os.path.splitext(url)[-1]  
        if ext not in [".jpg", ".jpeg", ".png", ".gif", ".bmp", ".webp"]:
            ext = ".jpg"  


        filename = f"image{index}{ext}"  
        file_path = os.path.join("images", filename)

        with open(file_path, "wb") as file:
            file.write(response.content)

        print(f"Saved: {file_path}")

    except requests.RequestException as e:
        print(f"Failed to download {url}: {e}")
```
