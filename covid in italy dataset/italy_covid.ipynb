```python
df['day'] =df['date'].dt.day
df['month'] = df['date'].dt.month
df['year'] = df['date'].dt.year
df['weekday'] = df['date'].dt.day_name()
df
```

```python
df[df['month'] == 5].groupby(['month'])[['new_cases','new_deaths' , 'new_tests']].sum()
```

```python
maxday_cases = df.groupby('day')['new_cases'].sum()
maxday_cases.sort_values(ascending=False)
print("The most day with new cases is : ",maxday_cases.idxmax())
```

```python
maxweekday_cases = df.groupby('weekday')['new_cases'].sum()
maxweekday_cases.sort_values(ascending=False)
print("The most weekday with new cases is : ",maxweekday_cases.idxmax())
```

```python
month_wise = df.groupby('month')[['new_cases','new_deaths','new_tests']].sum()
month_wise
```

```python
day_wise = df.['new_cases'].sum()
day_wise
```

```python
df.plot(x='date', y='new_cases', kind='line').grid(True)
```

```python
plt.figure(figsize=(10,6))
plt.plot(df['date'], df['new_cases'])
plt.xlabel('Date')
plt.ylabel('New Cases')
plt.title('Covid-19 Cases in Italy')
plt.grid()
plt.legend(['New Cases'])
plt.show()
```
